```python
import matplotlib.pyplot as plt
from sklearn import datasets, metrics

digits = datasets.load_digits()
```

```python
plt.imshow(digits.images[0], cmap =plt.cm.gray_r, interpolation='nearest')
```

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier #kNN

knn = KNeighborsClassifier(n_neighbors=6)#kNN객체생성

knn.fit(X_train, y_train) #학습

pred_knn = knn.predict(X_test) #예측값 저장

scores= metrics.plot_confusion_matrix(knn, X_test, y_test)
plt.show()

print(f"{metrics.classification_report(y_test, pred_knn)}\n")
```

```python
from sklearn.tree import DecisionTreeClassifier #DecisionTree

dt = DecisionTreeClassifier(random_state=0) #DecisionTree 객체생성

dt.fit(X_train, y_train) # DecisionTree 학습

pred_dt = dt.predict(X_test) #예측값 저장

scores= metrics.plot_confusion_matrix(dt, X_test, y_test)       #혼동행렬 출력
plt.show()

print(f"{metrics.classification_report(y_test, pred_dt)}\n")   #결과 출력
```

```python
from sklearn.ensemble import RandomForestClassifier  #RandomForest

rf = RandomForestClassifier(random_state=0) #RandomForest 객체생성

rf.fit(X_train, y_train) #Randomforest 학습

pred_rf = rf.predict(X_test) #예측값 저장

scores= metrics.plot_confusion_matrix(rf, X_test, y_test)       #혼동행렬 출력
plt.show()

print(f"{metrics.classification_report(y_test, pred_rf)}\n")   #결과 출력
```
