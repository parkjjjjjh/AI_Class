```python
#csv 파일을 불러들임. columns이름 있으니 따로 설정할 필요 없음
import pandas as pd
import numpy as np

df = pd.read_excel("C:/Users/Parkjunhan/Desktop/Raisin_Dataset/Raisin_Dataset/Raisin_Dataset.xlsx")
df
```

```python
# x,y 나누는 코드 

X=df.drop('Class',axis=1) #axis=0or1임. 1로하면 x축을 기준으로 나눔. 0은 y축을 기준으로 나눔. 
X.head()
```

```python
# x,y 나누는 코드 

y=df['Class']
y.value_counts()
```

```python
# x트레인,y트레인,x테스트,y테스트로 코드 나눠줌

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
# 라이브러리 import 

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# X_train, y_train으로 훈련 (밑에서부터 5개의 다른 모델을 적용해서 ML, 최소 Logisregression은 하기)

from sklearn.svm import SVC

clf_svm = SVC(random_state=0)    # SVM 모델
clf_svm.fit(X_train, y_train)    # fit 함수로 훈련

pred_svm = clf_svm.predict(X_test)  # 테스트값으로 훈련 잘했나 비교하기 위해 예측값 생성

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))     #예측값과 비교하는 코드(y테스트와 pred_svm과 비교)
print(confusion_matrix(y_test, pred_svm))   #혼동행렬 출력
```

```python
clf_lr = LogisticRegression(random_state=0)  #LogisticRegrssion 모델     *변수이름 바꿔보기
clf_lr.fit(X_train, y_train)  # fit 함수로 훈련

pred_lr = clf_lr.predict(X_test)  

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
clf_dt = DecisionTreeClassifier(random_state=0) #모델 결정
clf_dt.fit(X_train, y_train) #fit시킴

pred_dt = clf_dt.predict(X_test) #평가

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
clf_nn = MLPClassifier(random_state=0)
clf_nn.fit(X_train, y_train)

pred_nn = clf_nn.predict(X_test)

print ("\n--- Neural Network Classifier ---")
print (accuracy_score(y_test, pred_nn))
print (confusion_matrix(y_test, pred_nn))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print ("\n--- Radom Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```
